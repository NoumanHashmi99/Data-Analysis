```python
lst = ['[',']']
def clean(value):
    for i in lst:
        value = value.str.replace(i,'',regex=False)
    return value
```

```python
clean(df['Peak'][14])
```

```python
df['Peak']=clean(df['Peak'])
```

```python
df.iloc[:,[2,-1]] = df.iloc[:,[2,-1]].apply(clean)
```
